```python
#cell 4
from tensorflow.keras.preprocessing.image import ImageDataGenerator
IMG_SIZE  = (224, 224)
BATCH     = 32
BASE_PATH = '/content/drive/MyDrive/BellPepper_Project/dataset'
CLASSES   = ['Bacterial_Spot', 'Healthy']
train_aug = ImageDataGenerator(
    rotation_range=40,           # stronger rotation
    width_shift_range=0.25,      # stronger shift
    height_shift_range=0.25,
    shear_range=0.20,
    zoom_range=0.30,             # stronger zoom
    horizontal_flip=True,
    vertical_flip=True,          # added vertical flip
    brightness_range=[0.65, 1.35],  # stronger brightness
    channel_shift_range=40.0,    # stronger channel shift
    fill_mode='reflect'          # reflect instead of nearest
)
val_aug  = ImageDataGenerator()
test_aug = ImageDataGenerator()
train_gen = train_aug.flow_from_directory(
    f'{BASE_PATH}/train',
    target_size=IMG_SIZE,
    batch_size=BATCH,
    class_mode='binary',
    classes=CLASSES,
    seed=42,
    shuffle=True
)
val_gen = val_aug.flow_from_directory(
    f'{BASE_PATH}/val',
    target_size=IMG_SIZE,
    batch_size=BATCH,
    class_mode='binary',
    classes=CLASSES,
    seed=42,
    shuffle=False
)
test_gen = test_aug.flow_from_directory(
    f'{BASE_PATH}/test',
    target_size=IMG_SIZE,
    batch_size=1,
    class_mode='binary',
    classes=CLASSES,
    seed=42,
    shuffle=False
)
print("Class indices:", train_gen.class_indices)
print(f"Train: {train_gen.samples} | Val: {val_gen.samples} | Test: {test_gen.samples}")
```

```python
#cell 5
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras import layers, Model
import tensorflow.keras as keras

IMG_SIZE  = (224, 224)
BASE_PATH = '/content/drive/MyDrive/BellPepper_Project/dataset'
SAVE_PATH = '/content/drive/MyDrive/BellPepper_Project/saved_model'
import os; os.makedirs(SAVE_PATH, exist_ok=True)
def build_model():
  base = EfficientNetB0(
      include_top=False,
      weights='imagenet',
      input_shape=(224, 224, 3)
      )
  base.trainable = False
  inputs = keras.Input(shape=(224, 224, 3))
  x = base(inputs, training=False)
  x = layers.GlobalAveragePooling2D()(x)
  x = layers.BatchNormalization()(x)
  x = layers.Dropout(0.55)(x)
  x = layers.Dense(128, activation='relu',
                   kernel_regularizer=keras.regularizers.l2(1e-3))(x)
  x = layers.BatchNormalization()(x)
  x = layers.Dropout(0.45)(x)
  outputs = layers.Dense(1, activation='sigmoid')(x)
  model = Model(inputs, outputs, name='BellPepperCNN')
  model.compile(
      optimizer=keras.optimizers.Adam(1e-4),
      loss=tf.keras.losses.BinaryCrossentropy(
          label_smoothing=0.15
          ),
      metrics=[
          'accuracy',
          keras.metrics.AUC(name='auc'),
          keras.metrics.Precision(name='precision'),
          keras.metrics.Recall(name='recall')
          ]
      )
  return model, base
model, base = build_model()
model.summary()
print("✅ Model built!")
```

```python
#cell 6
from tensorflow.keras.callbacks import (
      ModelCheckpoint, EarlyStopping, ReduceLROnPlateau)
callbacks_p1 = [
    ModelCheckpoint(
        f'{SAVE_PATH}/best_phase1.keras',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
        ),
    EarlyStopping(
         monitor='val_accuracy',
         patience=5,              # stricter early stopping
         restore_best_weights=True,
         verbose=1
         ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.4,
        patience=3,
        min_lr=1e-7,
        verbose=1
        ),
]
print("PHASE 1: Training head only...")
history1 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=20,
    callbacks=callbacks_p1,
    verbose=1
)
print(f"✅ Phase 1 best val_accuracy: {max(history1.history['val_accuracy']):.4f}")
```

```python
#cell 7
# Unfreeze only TOP 30 layers — less fine-tuning = less overfitting
base.trainable = True
for layer in base.layers[:-30]:
    layer.trainable = False

# Count trainable layers
trainable = sum([1 for l in base.layers if l.trainable])
print(f"Trainable layers in base: {trainable}")

model.compile(
    optimizer=keras.optimizers.Adam(5e-6),  # very small LR
    loss=tf.keras.losses.BinaryCrossentropy(
        label_smoothing=0.15
        ),
        metrics=[
            'accuracy',
            keras.metrics.AUC(name='auc'),
            keras.metrics.Precision(name='precision'),
            keras.metrics.Recall(name='recall')
            ]
    )

callbacks_p2 = [
    ModelCheckpoint(
        f'{SAVE_PATH}/bell_pepper_final.keras',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
        ),
    EarlyStopping(
        monitor='val_accuracy',
        patience=8,
        restore_best_weights=True,
        verbose=1
        ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.4,
        patience=4,
        min_lr=1e-8,
        verbose=1
        ),
]
print("PHASE 2: Fine-tuning top 30 layers...")
history2 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=50,
    initial_epoch=len(history1.history['loss']),
    callbacks=callbacks_p2,
    verbose=1
)
print(f"✅ Phase 2 best val_accuracy: {max(history2.history['val_accuracy']):.4f}")
```

```python
#cell 8
results = model.evaluate(test_gen, verbose=1)
metrics = dict(zip(model.metrics_names, results))
print('\n=== TEST SET METRICS ===')
for k, v in metrics.items():
    print(f'  {k:<12}: {v:.4f}')

# ── Plot Training Curves ─────────────────────────────────────
import matplotlib.pyplot as plt

acc  = history1.history['accuracy']  + history2.history['accuracy']
vacc = history1.history['val_accuracy'] + history2.history['val_accuracy']
loss = history1.history['loss']      + history2.history['loss']
vloss= history1.history['val_loss']  + history2.history['val_loss']
phase1_end = len(history1.history['loss'])

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Bell Pepper CNN — Training Metrics', fontsize=14)

axes[0].plot(acc,  label='Train Acc',  color='#2ecc71')
axes[0].plot(vacc, label='Val Acc',    color='#e74c3c', linestyle='--')
axes[0].axvline(phase1_end-1, color='gray', linestyle=':', label='Fine-tune')
axes[0].set_title('Accuracy'); axes[0].legend(); axes[0].set_ylim([0,1])

axes[1].plot(loss,  label='Train Loss', color='#3498db')
axes[1].plot(vloss, label='Val Loss',   color='#e67e22', linestyle='--')
axes[1].axvline(phase1_end-1, color='gray', linestyle=':')
axes[1].set_title('Loss'); axes[1].legend()

plt.tight_layout()
plt.savefig('/content/drive/MyDrive/BellPepper_Project/training_curves.png', dpi=150)
plt.show()
print('Curves saved to Drive!')
```

```python
#cell 11
import tensorflow as tf
import cv2, numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import glob
import os
BASE_PATH = '/content/drive/MyDrive/BellPepper_Project/dataset'
print("test subfolders:", os.listdir(f'{BASE_PATH}/test'))


def get_gradcam(model, img_array):
  efficientnet = [l for l in model.layers if 'efficientnet' in l.name][0]
  target_layer = efficientnet.get_layer('top_conv')
  grad_model = tf.keras.Model(
      inputs=efficientnet.input,
      outputs=[target_layer.output, efficientnet.output])
  img_tensor = tf.cast(img_array, tf.float32)
  with tf.GradientTape() as tape:
    tape.watch(img_tensor)
    conv_out, predictions = grad_model(img_tensor)
    loss = predictions[:, 0]
  grads = tape.gradient(loss, conv_out)
  pooled = tf.reduce_mean(grads, axis=(0, 1, 2))
  heatmap = conv_out[0] @ pooled[..., tf.newaxis]
  heatmap = tf.squeeze(heatmap)
  heatmap = tf.maximum(heatmap, 0).numpy()
  heatmap = (heatmap - heatmap.min()) / (heatmap.max() - heatmap.min() + 1e-8)
  return heatmap

def overlay_disease(img_path, heatmap, threshold=0.5, alpha=0.65):
  """Bacterial Spot — red hotspots only on lesion areas"""
  img = cv2.imread(img_path)
  img = cv2.resize(img, (224, 224))
  h   = cv2.resize(heatmap, (224, 224))
  h_thresh = np.where(h >= threshold, h, 0)
  colormap = plt.get_cmap('jet')(h_thresh)[:,:,:3]
  heatmap_bgr = np.uint8(255 * colormap[:,:,::-1])
  mask    = (h_thresh > 0).astype(np.uint8)
  mask_3ch = np.stack([mask]*3, axis=-1)
  blended = np.where(mask_3ch,
                     cv2.addWeighted(img, 1-alpha, heatmap_bgr, alpha, 0),
                     img)
  return blended
def overlay_healthy(img_path, heatmap, alpha=0.55):
  """Healthy — cool blue/green overlay, NO red/yellow"""
  img = cv2.imread(img_path)
  img = cv2.resize(img, (224, 224))
  h   = cv2.resize(heatmap, (224, 224))
  # Invert heatmap so healthy areas show cool colors
  h_inverted = 1.0 - h
  # Use 'cool' colormap — only blues and cyans, no red/yellow
  colormap = plt.get_cmap('cool')(h_inverted)[:,:,:3]
  heatmap_bgr = np.uint8(255 * colormap[:,:,::-1])
  blended = cv2.addWeighted(img, 1-alpha, heatmap_bgr, alpha, 0)
  return blended
def get_images(folder, n=3):
  imgs = (glob.glob(f'{folder}/*.jpg') +
          glob.glob(f'{folder}/*.JPG') +
          glob.glob(f'{folder}/*.jpeg') +
          glob.glob(f'{folder}/*.png'))
  return imgs[:n]
bact_imgs    = get_images(f'{BASE_PATH}/test/Bacterial_Spot')
healthy_imgs = get_images(f'{BASE_PATH}/test/Healthy')
# ── Bacterial Spot — red hotspots ────────────────────────────
fig1, axes1 = plt.subplots(len(bact_imgs), 2, figsize=(10, 5*len(bact_imgs)))
fig1.suptitle('Bacterial Spot — Grad-CAM Disease Regions', fontsize=13)
for i, path in enumerate(bact_imgs):
  img   = tf.keras.utils.load_img(path, target_size=(224,224))
  arr   = tf.keras.utils.img_to_array(img)
  batch = np.expand_dims(arr, 0)
  pred     = float(model.predict(batch, verbose=0)[0][0])
  severity = round(1 - pred, 4)
  heatmap = get_gradcam(model, batch)
  h_resized = cv2.resize(heatmap, (224, 224))

  axes1[i][0].imshow(cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB))
  axes1[i][0].set_title(f'Ground Truth: Bacterial Spot\nScore: {severity}')
  axes1[i][0].axis('off')

  im = axes1[i][1].imshow(h_resized, cmap='jet', vmin=0, vmax=1)
  axes1[i][1].set_title('Grad-CAM')
  axes1[i][1].axis('off')
plt.colorbar(im, ax=axes1[i][1], fraction=0.046, pad=0.04)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/BellPepper_Project/gradcam_bacterial.png', dpi=150)
plt.show()
# ── Healthy — cool blue heatmap, no highlights ────────────────
fig2, axes2 = plt.subplots(len(healthy_imgs), 2, figsize=(10, 5*len(healthy_imgs)))
fig2.suptitle('Healthy Leaves — Grad-CAM (No Disease)', fontsize=13)
for i, path in enumerate(healthy_imgs):
  img   = tf.keras.utils.load_img(path, target_size=(224,224))
  arr   = tf.keras.utils.img_to_array(img)
  batch = np.expand_dims(arr, 0)
  pred     = float(model.predict(batch, verbose=0)[0][0])
  severity = round(pred, 4)
  heatmap = get_gradcam(model, batch)
  overlay = overlay_healthy(path, heatmap)
  axes2[i][0].imshow(cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB))
  axes2[i][0].set_title(f'Original | Healthy | Score: {severity}')
  axes2[i][0].axis('off')
  axes2[i][1].imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
  axes2[i][1].set_title('Grad-CAM — Blue = Healthy (No Lesion)')
  axes2[i][1].axis('off')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/BellPepper_Project/gradcam_healthy.png', dpi=150)
plt.show()
print("✅ Both saved to Drive!")
```
